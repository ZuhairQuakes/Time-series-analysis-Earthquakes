```python
data_T_UP = data_UP.loc[data_UP['TYPE'] == 'THRUST']
data_SS_UP = data_UP.loc[data_UP['TYPE'] == 'SS']
data_N_UP = data_UP.loc[data_UP['TYPE'] == 'NORM']
```

```python
data_df = data_T_UP.sort_values(by = 'ALONG',ascending=True)
data_along_pos_T = pd.unique(data_df['ALONG'])
data_along_energy_T = []

grp = data_df.groupby('ALONG')

for pos in data_along_pos_T:
    to_proc = grp.get_group(pos) 
    exp = np.asarray(to_proc['EX'])
    mo = np.asarray(to_proc['MO'])
    energy = mo*np.power(10., exp)
    data_along_energy_T.append(np.sum(energy))
    
data_along_energy_T = np.asarray(data_along_energy_T)
```

# Cumulative energy in the downgoing plate and upper plate and both

```python
data_UP = data[(data['P(km)'] > 100) & (data['P(km)'] < 700)] # towards landwards
data_UP = data_UP.loc[data['CLASS'] == 'UPPER']
data_T_UP = data_UP.loc[data['TYPE'] == 'THRUST']
data_T_UP = data_T_UP[(data_T_UP['DATE'] > '2004-12-25')]


data_in_slab = data[(data['P(km)'] > 0) & (data['DEPTH'] < 600)] # only in the subducting slab
data_DW = data_in_slab.loc[data['CLASS'] == 'DGOING']

data_MT = data.loc[data['CLASS'] == 'MTHRUST']
```

```python
# code for getting the cumulative energy in the downgoing and upper plate and both

ds = 100 # steps in the along direction (200 km for this case)
s_max = data['ALONG'].max() # earthquake data

s_arr = np.arange(0, s_max + ds, ds)
s_idx = np.arange(0, s_arr.shape[0], 1) #index for iterating through s_arr
```

```python
cum_energy_up = np.zeros(s_arr.shape[0])
cum_energy_dw = np.zeros(s_arr.shape[0])
cum_energy_mt = np.zeros(s_arr.shape[0])
```

# Excluding Megathrust energy release from the total energy

```python
# code for getting the cumulative value per bin without using grouby
for i, pos in zip(s_idx, s_arr):
    
    subset_up = data_T_UP[data_T_UP['ALONG'] == pos]
    subset_dw = data_DW[data_DW['ALONG'] == pos]
#     subset_mt = data_MT[data_MT['ALONG'] == pos]
    
#     subset_up = data_T_UP[data_T_UP['ALONG'] == pos]
#     subset_dw = data_T_DW[data_T_DW['ALONG'] == pos]
    
    # work with upper
    exp = np.asarray(subset_up['EX'])
    mo = np.asarray(subset_up['MO'])
    energy = mo*np.power(10., exp)
    cum_energy_up[i] = energy.sum() #total energy in this bin
    
    # work with downgoing
    exp = np.asarray(subset_dw['EX'])
    mo = np.asarray(subset_dw['MO'])
    energy = mo*np.power(10., exp)
    cum_energy_dw[i] = energy.sum() # total energy in this bin
    
# data_along_pos = pd.unique(data_df['ALONG'])    
 
# get total moment per bin by adding cum_energy_dw and cum_energy_up
cum_energy_tot = cum_energy_dw + cum_energy_up
z = cum_energy_up / cum_energy_tot
# z2 = cum_energy_dw / cum_energy_tot
```

```python
fig, ax1 = plt.subplots(dpi = 300)

# plt.yscale('log')
color = 'tab:gray'
ax1.set_xlabel('Distance along trench [km]')
ax1.set_ylabel('Z number', color=color)  # we already handled the x-label with ax1
# plot = ax1.plot(s_arr, cum_energy_tot, color = 'lightskyblue', linewidth = 0.5, edgecolor = 'k', width = width, alpha = 0.5, label= 'Seismic Moment Release')
plot = ax1.bar(s_arr, z, color = 'r', alpha = 0.5, edgecolor = 'k', width = 100)
# plot = ax1.plot(s_arr, z1, color = 'r', linewidth = 1, label = 'UP_T/total')
# plot = ax1.errorbar(s_arr, z1, xerr = 200, color='r')
# plot = ax1.plot(s_arr, z2, color = 'lightskyblue', linewidth = 1, label = 'DW/total')
# plot = ax1.plot(s_arr, cum_energy_tot, color = 'g', linewidth = 2, label = 'Total')
# plt.legend(loc = 2)

# Smoothing the curve (https://www.geeksforgeeks.org/how-to-plot-a-smooth-curve-in-matplotlib/)
# X_Y_Spline = make_interp_spline(s_arr, z1)
# X_ = np.linspace(s_arr.min(), s_arr.max(), 300)
# Y_ = X_Y_Spline(X_)

# plot = ax1.plot(X_, Y_, color = color, linewidth = 1)


ax1.tick_params(axis='y', labelcolor=color)
# ax1.set_ylim([-0.2e21, 1e21])
# plt.legend()

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:gray'
ax2.set_ylabel('Slab depth [km]', color=color)
ax2.plot(slab_length['ALONG'], slab_length['LENGTH']*-1, '-', color=color, alpha = 0.8, markersize = 3, linewidth = 2)
ax2.tick_params(axis='y', labelcolor=color)
# ax2.set_ylim([-30, 20])
ax2.set_xlim([0, 5500])
# plt.title('Southeast Asia - Ratio')
# plt.savefig('SU_UP_T/total_DW+UP.png', dpi = 300)
```

# Including Megathrust

```python
# code for getting the cumulative value per bin without using grouby
for i, pos in zip(s_idx, s_arr):
    
    subset_up = data_T_UP[data_T_UP['ALONG'] == pos]
    subset_dw = data_DW[data_DW['ALONG'] == pos]
    subset_mt = data_MT[data_MT['ALONG'] == pos]
    
#     subset_up = data_T_UP[data_T_UP['ALONG'] == pos]
#     subset_dw = data_T_DW[data_T_DW['ALONG'] == pos]
    
    # work with upper
    exp = np.asarray(subset_up['EX'])
    mo = np.asarray(subset_up['MO'])
    energy = mo*np.power(10., exp)
    cum_energy_up[i] = energy.sum() #total energy in this bin
    
    # work with downgoing
    exp = np.asarray(subset_dw['EX'])
    mo = np.asarray(subset_dw['MO'])
    energy = mo*np.power(10., exp)
    cum_energy_dw[i] = energy.sum() # total energy in this bin
    
     # work with Megathrust
    exp = np.asarray(subset_mt['EX'])
    mo = np.asarray(subset_mt['MO'])
    energy = mo*np.power(10., exp)
    cum_energy_mt[i] = energy.sum() # total energy in this bin
    
# data_along_pos = pd.unique(data_df['ALONG'])    
 
# get total moment per bin by adding cum_energy_dw and cum_energy_up
cum_energy_tot = cum_energy_dw + cum_energy_up + cum_energy_mt
z = cum_energy_up / cum_energy_tot
# z2 = cum_energy_dw / cum_energy_tot
```
